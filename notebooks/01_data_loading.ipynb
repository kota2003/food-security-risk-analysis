## 0. Setup

```python
import pandas as pd
import numpy as np
import os

# Display settings
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 50)
pd.set_option('display.float_format', '{:.2f}'.format)

# Data paths
RAW_DATA_PATH = '../data/raw/'

print('Setup complete.')
print(f'Files in raw data folder:')
for f in os.listdir(RAW_DATA_PATH):
    print(f'  {f}')
```

---
## 1. FAOSTAT — Suite of Food Security Indicators

```python
fao_fs = pd.read_csv(RAW_DATA_PATH + 'Suite_of_Food_Security_Indicators.csv')

print('=== Shape ===')
print(fao_fs.shape)

print('\n=== Columns ===')
print(fao_fs.columns.tolist())

print('\n=== Head ===')
fao_fs.head()
```

```python
# Unique items (indicators)
print('=== Unique Items (Indicators) ===')
print(fao_fs['Item'].unique())

print('\n=== Year Range ===')
print(f"Min: {fao_fs['Year'].min()}, Max: {fao_fs['Year'].max()}")

print('\n=== Number of Countries ===')
print(fao_fs['Area'].nunique())
```

```python
# Missing values
print('=== Missing Values ===')
print(fao_fs.isnull().sum())

print('\n=== Missing Rate for Value column ===')
missing_rate = fao_fs['Value'].isnull().sum() / len(fao_fs) * 100
print(f'{missing_rate:.1f}%')
```

## 4. World Bank — World Development Indicators

```python
# World Bank uses '..' for missing values
wb = pd.read_csv(
    RAW_DATA_PATH + 'WorldDevelopmentIndicators.csv',
    na_values=['..']  # convert '..' to NaN
)

print('=== Shape ===')
print(wb.shape)

print('\n=== Columns ===')
print(wb.columns.tolist())

print('\n=== Head ===')
wb.head()
```

```python
# Unique series (indicators)
print('=== Unique Series (Indicators) ===')
print(wb['Series Name'].unique())

print('\n=== Number of Countries ===')
print(wb['Country Name'].nunique())
```

```python
# Missing rate per indicator
print('=== Missing Rate per Indicator ===')
year_cols = [c for c in wb.columns if 'YR' in c]
for series in wb['Series Name'].dropna().unique():
    subset = wb[wb['Series Name'] == series][year_cols]
    missing = subset.isnull().sum().sum()
    total = subset.size
    print(f'{str(series)[:50]}: {missing/total*100:.1f}% missing')
```
